### Input-'person_age', 'person_income', 'person_home_ownership','person_emp_length', 
### 'loan_intent', 'loan_grade', 'loan_amnt','loan_int_rate', 'loan_percent_income',
### 'cb_person_default_on_file', 'cb_person_cred_hist_length'.
### Target-'loan_status'

### Alg-classification task
### Evaluation- Accuracy
### Split the dataset into Training and Testing (recommended 75:25 split).

```python
y=df['loan_status']
X=df[[ 'person_age', 'person_income', 'person_home_ownership','person_emp_length',
'loan_intent', 'loan_grade', 'loan_amnt','loan_int_rate', 'loan_percent_income',
'cb_person_default_on_file', 'cb_person_cred_hist_length'
      ]]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.75,random_state=100)
print(X_train.shape,y_train.shape)
print(X_test.shape,y_test.shape)
```

```python
X_train.head()
```

```python
X_train_cat=X_train.select_dtypes(include=['object'])
X_train_cat.head()
```

```python
X_train_num=X_train.select_dtypes(include=['int64','float64'])
X_train_num.head()
```

# Data preparation on train data:

```python
from sklearn.preprocessing import StandardScaler
Scaler=StandardScaler()
x_train_num_rescaled=pd.DataFrame(Scaler.fit_transform(X_train_num),
                                 columns=X_train_num.columns,
                                 index=X_train_num.index)
x_train_num_rescaled.head()
```

```python
X_train_num.describe()
```

```python
X_train_cat['person_home_ownership'].value_counts(normalize=True)
```

```python
X_train_cat['loan_intent'].value_counts(normalize=True)
```

```python
X_train_cat['loan_grade'].value_counts(normalize=True)
```

```python
X_train_cat['cb_person_default_on_file'].value_counts(normalize=True)
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder( drop='first',sparse=False)

X_train_cat_ohe = pd.DataFrame(encoder.fit_transform(X_train_cat), 
                               columns=encoder.get_feature_names_out(X_train_cat.columns), 
                               index = X_train_cat.index)

X_train_cat_ohe.head()
```

```python
encoder.categories_
```

```python
X_train_transformed=pd.concat([x_train_num_rescaled,X_train_cat_ohe],axis=1)
X_train_transformed.head()
```

# Data preparation on test data:

```python
X_test_num=X_test.select_dtypes(include=['int64','float64'])
X_test_num.head()
```

```python
x_test_num_rescaled=pd.DataFrame(Scaler.transform(X_test_num),
                                 columns=X_test_num.columns,
                                 index=X_test_num.index)
x_test_num_rescaled.head()
```

```python
X_test_cat=X_test.select_dtypes(include=['object'])
X_test_cat.head()
```

```python
X_test_cat_le = pd.DataFrame(encoder.transform(X_test_cat), 
                               columns=encoder.get_feature_names_out(X_test_cat.columns), 
                               index = X_test_cat.index)

X_test_cat_le.head()
```

```python
X_test_transformed=pd.concat([x_test_num_rescaled,X_test_cat_le],axis=1)
X_test_transformed.head()
```

# LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train_transformed, y_train)
```

```python
y_test_pred = classifier.predict(X_test_transformed)
```

```python
from sklearn import metrics

metrics.accuracy_score(y_test, y_test_pred)
```

# KNeighborsClassifier

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier()
classifier.fit(X_train_transformed, y_train)

y_test_pred = classifier.predict(X_test_transformed)

metrics.accuracy_score(y_test, y_test_pred)
```

# DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier()
classifier.fit(X_train_transformed, y_train)

y_test_pred = classifier.predict(X_test_transformed)

metrics.accuracy_score(y_test, y_test_pred)
```

# RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier()
classifier.fit(X_train_transformed, y_train)

y_test_pred = classifier.predict(X_test_transformed)

metrics.accuracy_score(y_test, y_test_pred)
```

# Support Vector Machines

```python
from sklearn.svm import SVC
classifier=SVC()
classifier.fit(X_train_transformed, y_train)
y_test_pred = classifier.predict(X_test_transformed)
metrics.accuracy_score(y_test, y_test_pred)
```

# conclusion:
* Random forest classification has the highest accuracy for prediction.
* when compared to other Classifiers Random forest classification having more accuracy.
* accuracy value for Random forest classification is 0.92.
